```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

$$
y = x^2
$$

```python
x = np.linspace(-10, 10, 400)
y = np.linspace(0, 10, 400)
x, y = np.meshgrid(x, y)

plt.contour(x, y, (y - x**2), [0], colors='k')

plt.show()
```

```python
x_range = np.arange(-1*np.pi,1*np.pi,0.1)

def f_a(x):
    return math.sin(math.pi * x)

def f_b(x):
    return x**2 - 1

def f_c(x):
    return f_a(x) / f_b(x)

a = [f_a(x) for x in x_range]
b = [f_b(x) for x in x_range]
c = [f_c(x) for x in x_range]

plt.plot(x_range,a,x_range,b,x_range,c)
plt.show()
```

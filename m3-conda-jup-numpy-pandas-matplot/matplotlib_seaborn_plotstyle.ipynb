# Matplotlib & Seaborn Plot Styling


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

pokemon = pd.read_csv('./data/pokemon.csv')

x = pokemon['generation_id'].unique()
y = pokemon['generation_id'].value_counts(sort=False)


# allow charts to be drawn inside this notebook
%matplotlib inline
```

## Figures & Axes

The base of visualization in matplotlib is a [**Figure**](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html) object. Contained within each Figure will be one or more [**Axes**](https://matplotlib.org/api/axes_api.html) objects, each Axes object containing a number of other elements that represent each plot.


```python
# create a new Figure
fig = plt.figure(figsize=(5,3))


# The argument of add_axes represents the dimensions [left, bottom, width, height] of the new axes. 
# All quantities are in fractions of figure width and height.
ax = fig.add_axes([.125, .125, .775, .755])
ax.hist(data=pokemon, x='speed')

fig.show()
```

## Subplots


```python
# Resize the chart, and have two plots side-by-side
# set a larger figure size for subplots
plt.figure(figsize = [20, 5]) 

# create shared bins
a_max = pokemon['attack'].max()
d_max = pokemon['defense'].max()
s_max = pokemon['speed'].max()

bin_max = max((a_max, d_max, s_max))
bins = np.arange(0, bin_max+5, 5)

plt.subplot(1, 3, 1) # 1 row, 3 cols, subplot 1
plt.hist(data = pokemon, x = 'attack', bins = bins, color='red')
plt.xlabel('Attack')

plt.subplot(1, 3, 2) # 1 row, 3 cols, subplot 2
plt.hist(data = pokemon, x = 'defense', bins = bins, color='blue')
plt.xlabel('Defence')

plt.subplot(1, 3, 3) # 1 row, 3 cols, subplot 3
plt.hist(data = pokemon, x = 'speed', bins = bins, color='orange')
plt.xlabel('Speed')

plt.show()
```

## Labels & Ticks

```python
base_color = sb.color_palette()[2]

gen_ticks = ["Gen %d"%(gen) for gen in x]

plt.bar(x, y)
plt.xlabel('Generation')
plt.ylabel('Count')
plt.xticks(x, labels = gen_ticks)

plt.show()
```
